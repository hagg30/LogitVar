```python
import os.path 

datasets = ["LostAndFound", "Static"]
methods = ["dnlnet", "ocrnet", "ISANet", "DANet", "DeepLabV3Plus", "DeepLabV3"]
subcat = ["max", "var", "sum", "maxex", "varex", "sumex"]
print_cat = ["SML", "LOV", "Sum", "SML + IBH", "LOV + IBH", "Sum + IBH"]
score_met = ["\multirow{6}{*}{80.41}", "\multirow{6}{*}{81.35}", "\multirow{6}{*}{80.81}", "\multirow{6}{*}{80.61}", "\multirow{6}{*}{80.52}", "\multirow{6}{*}{80.10}"]
front_met = ["\multirow{6}{*}{DNLNet \cite{yin2020disentangled}}", "\multirow{6}{*}{OCRNet \cite{yuan2020object}}", "\multirow{6}{*}{ISANet \cite{huang2019interlaced}}", 
            "\multirow{6}{*}{DANet \cite{fu2019dual}}", "\multirow{6}{*}{DeepLabV3+ \cite{chen2018encoder}}", "\multirow{6}{*}{DeepLabV3 \cite{chen2017rethinking}}"]

for idx_m, m in enumerate(methods):
    print(front_met[idx_m])
    for idx, c in enumerate(subcat):
        s3 = print_cat[idx]
        for idx_d, d in enumerate(datasets):
            #print(os.path.isfile("./"+d+"/"+m+c))
            with open(("./"+d+"/"+m+c), "r") as infile:
                words = infile.read().split(" ")
                s3 += ' & {ap:.2f} & {fpr:.2f} '.format(ap=float(words[1])*100, fpr=float(words[3])*100)
        if idx == 0:
            s3 += "&" + score_met[idx_m]
            print( "&" + s3 + "\\\\" )
        else:
            print( "&" + s3 + "\\\\" )
    print("\hline")
    print()
#
```
